```python
# set the oauth parameters for the client request
oauth_consumer_key = '3cf2ef5fe4c5f03864af5210991144ec96736087d4e2d32cce613d27c941bfe7'
oauth_token = 'c69eda6bb5ede3e6858807950213992ece347f7e2844dc2182a9e569da1826bc'
oauth_consumer_secret = 'eebf4fe840f185ebbea477ae6bdbc32e5f51c9d4cf86c8402a1e5fef89997f4e'
oauth_token_secret = '0840215b4d37133aecd7fca288d2ec12c1df3b8d716830394e9e52c69d54d02a'
oauth_signature_method = 'HMAC-SHA256'
oauth_timestamp = str(round(time.time()))
oauth_version = '1.0'
url = 'https://1238277-sb1.suitetalk.api.netsuite.com/services/rest/record/v1/customer'
account_id = '1238277_SB1'
http_method = 'GET'

# create an 11 character alpha-numeric nonce
def generate_nonce():

  # create a list of all possible characters for the nonce and append them to the nonce list
  characters = string.ascii_letters + string.digits
  nonce = []
  for _ in range(11):
    nonce.append(random.choice(characters))

  return "".join(nonce)

oauth_nonce = generate_nonce()

# create a signature using a base string of a consistent reproducible concatenation of several of the http request elements into a single string
def oath_signature(oauth_consumer_key, oauth_token, oauth_signature_method, oauth_timestamp, oauth_nonce, oauth_version, oauth_consumer_secret, oauth_token_secret):

  # set the base string to include all parameters for the request
  base_string = urllib.parse.quote("&".join([
      "oauth_consumer_key=" + oauth_consumer_key,
      "oauth_nonce=" + oauth_nonce,
      "oauth_signature_method=" + oauth_signature_method,
      "oauth_timestamp=" + oauth_timestamp,
      "oauth_token=" + oauth_token,
      "oauth_version=" + oauth_version
      ]), safe='')

  # use the http method, url, and base string to set the signing value
  signing_value = "&".join([
      http_method.upper(),
      urllib.parse.quote(url, safe=''),
      base_string
      ])

  # set the signature key
  signing_key = urllib.parse.quote("&".join([
      oauth_consumer_secret,
      oauth_token_secret
      ]), safe='&')

  # use the signing key and signing value to generate the oauth signature
  oauth_signature = urllib.parse.quote(
      base64.b64encode(
        hmac.new(
            signing_key.encode("utf-8"),
            signing_value.encode("utf-8"),
            hashlib.sha256
        ).digest()
      ).decode("utf-8"), safe='~()*!.\'')

  return oauth_signature

# create the headers
oauth_header = f'OAuth realm="{account_id}",oauth_consumer_key="{oauth_consumer_key}",oauth_token="{oauth_token}",oauth_signature_method="{oauth_signature_method}",oauth_timestamp="{oauth_timestamp}",oauth_nonce="{oauth_nonce}",oauth_version="{oauth_version}",oauth_signature="{oath_signature(oauth_consumer_key, oauth_token, oauth_signature_method, oauth_timestamp, oauth_nonce, oauth_version, oauth_consumer_secret, oauth_token_secret)}"'
headers = {
    "Authorization": oauth_header,
    "Cookie": "NS_ROUTING_VERSION=LAGGING"
    }

# make the http request
response = requests.request(http_method, url, headers=headers)
response
```
